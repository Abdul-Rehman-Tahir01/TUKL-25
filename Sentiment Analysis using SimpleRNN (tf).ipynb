```python
# Importing Libraries
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Loading the built-in IMDb review dataset
(X_train, Y_train), (X_test, Y_test) = tf.keras.datasets.imdb.load_data()

print('Training data shape: ', X_train.shape, ', ', Y_train.shape)
print('Test data shape: ', X_test.shape, ', ', Y_test.shape)
```

```python
print(X_train[0])
print(X_test[250])
```

-> This shows that the data in the dataset is already integer encoded. Let's check if the length of every vector is same or not.

```python
print(len(X_train[0]), len(X_train[1]), len(X_train[2]))
print(len(X_test[0]), len(X_test[1]), len(X_test[2]))
```

-> Clearly, the length is not the same. So, we have to zero pad all the vectors to make them the same length.

```python
# Applying zero padding
X_train = tf.keras.utils.pad_sequences(X_train, padding='post', maxlen=100)
X_test = tf.keras.utils.pad_sequences(X_test, padding='post', maxlen=100)

print('Training data shape: ', X_train.shape, ', ', Y_train.shape)
print('Test data shape: ', X_test.shape, ', ', Y_test.shape)
```

-> Now every vector in the training and test data has same length of 2494 . We can verify by again checking the length.

```python
print(len(X_train[0]), len(X_train[1]), len(X_train[2]))
print(len(X_test[0]), len(X_test[1]), len(X_test[2]))
```

```python
# Checking the data after padding
print(X_train[250])
print(X_test[0])
```

```python
# Checking the labels
print(Y_train[0], Y_train[1], Y_train[2])
print(Y_test[0], Y_test[1], Y_test[2])
```

-> This shows the sentiment of the sentences. '1' means positive sentiment, '0' means negative sentiment.

## <b>Using Learned Embedding</b>

```python
# Getting the vocabulary size
word_index = tf.keras.datasets.imdb.get_word_index()
vocab_size = len(word_index) + 1
print("Vocabulary size:", vocab_size)
```

```python
# Building the RNN

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=64, input_shape=(100,)),
    tf.keras.layers.SimpleRNN(32, return_sequences=False, activation='tanh', dropout=0.5),
    tf.keras.layers.Dense(1, activation='sigmoid', kernel_regularizer=tf.keras.regularizers.l2(0.1))
])

model.summary()
```

-> 88585x64 = 5,669,440 <br>
(64x32 + 32) + (32x32) = 3104 <br>
32x1 + 1 = 33

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(
   X_train, Y_train,
   epochs = 10,
   validation_data = (X_test, Y_test)
)  # Batch size is by default 32.
```

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

axs[0].plot(history.history['loss'], label='Train Loss')
axs[0].plot(history.history['val_loss'], label='Validation Loss')
axs[0].set_title('Model Loss')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Loss')
axs[0].legend()
axs[0].grid(True)

axs[1].plot(history.history['accuracy'], label='Train Accuracy')
axs[1].plot(history.history['val_accuracy'], label='Validation Accuracy')
axs[1].set_title('Model Accuracy')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Accuracy')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

-> Train accuracy is very greaster than test accuracy. This is case of overfitting.

```python
# Making a prediction
sample_review = ["I really enjoyed this movie. The plot was interesting, the characters were well-developed, and the pacing was just right. I found myself caring about the outcome and emotionally invested. Highly recommend!"]
print(sample_review)
```

```python
# Tokenizing for integer encoding first
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(sample_review)
tokenizer.word_index
```

```python
sample_review = tokenizer.texts_to_sequences(sample_review)
sample_review[0]
```

```python
# Padding to make it to the input shape that RNN expects
sample_review = tf.keras.utils.pad_sequences(sample_review, padding='post', maxlen=100)
print('Length of review: ', len(sample_review))
print(sample_review)
```

```python
input_review = np.expand_dims(sample_review[0], axis=0)
print(input_review.shape)
```

```python
pred = model.predict(input_review)
print("Prediction Prob: ", pred)


if pred[0] > 0.5:
    print("Sentiment: Positive")
else:
    print("Sentiment: Negative")
```
